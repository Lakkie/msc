# 1.数学分析 Mathematical Analysis

数学分析主要由微积分和级数理论组成，它所研究的主要对象是实函数，即以实数为自变量并且在实数中取值的函数。

‌数学分析‌（Mathematical Analysis）是数学的核心分支，以 ‌极限理论‌ 为基石，系统研究函数、序列、级数、微分和积分的性质及其应用。它通过严格的逻辑体系（如 ‌ε-δ epsilon-delta语言‌），将微积分学从直观的几何与物理描述上升为抽象的理论框架，并为现代数学的诸多领域（如实分析、泛函分析、微分方程等）提供基础工具。
